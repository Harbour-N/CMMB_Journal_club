Andrea Hawkins-Daarud et al. Quantifying Uncertainty and Robustness in a Biomathematical Model–Based Patient-Specific Response Metric for Glioblastoma. JCO Clin Cancer Inform 3, 1-8(2019).
DOI:10.1200/CCI.18.00066

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Laod in data
df = pd.read_csv('ds_cci.18.00066.csv')
```

```python
min(df['First_date_T1DG_Max'])
```

```python
colormap = plt.cm.nipy_spectral
colors = [colormap(i) for i in np.linspace(0, 1, len(df))]
# plot the data
fig, ax = plt.subplots()


for i in range(len(df)):
    ax.plot([0, df['dt_1'].iloc[i]], [df['First_date_T1DG_Min'].iloc[i], df['Second_data_T1Gd_Min'].iloc[i]], marker='o', color=colors[i])
    ax.plot([0, df['dt_1'].iloc[i]], [df['First_date_T1DG_Max'].iloc[i], df['Second_data_T1Gd_Max'].iloc[i]], marker='*', color=colors[i])
  
```
